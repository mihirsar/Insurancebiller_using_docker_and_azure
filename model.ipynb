```python
from pycaret.datasets import get_data

# Load the insurance dataset
insurance_data = get_data('insurance')
print(insurance_data)
```

```python
r_setup = setup(data=insurance_data, target='charges', session_id=123)
```

```python
best_model = compare_models(fold=10, sort='MAE')  # Sort models based on Mean Absolute Error (MAE)
```

```python
gbrmodel = create_model('gbr')
```

```python
tuned_gb_model = tune_model(gbrmodel, optimize='MAE', fold=5)
```

```python
# Evaluate the tuned model
evaluate_model(tuned_gb_model)
```

```python
predictions = predict_model(tuned_gb_model)

# Debugging: check the structure of the predictions DataFrame
print(predictions.head())

# Check if the 'Label' column exists in the DataFrame
if 'Label' in predictions.columns:
    # Calculate metrics manually
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    y_true = predictions['charges']
    y_pred = predictions['Label']

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("R-squared (R2):", r2)
else:
    print("Error: 'Label' column not found in predictions DataFrame.")
```
